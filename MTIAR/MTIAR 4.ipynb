```python
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy import stats
```

```python
data_1 = pd.read_csv('data_1.csv') 
data_2 = pd.read_csv('data_2.csv') 
```

```python
plt.plot(data_1['Agression'], data_1['IQ'], 'co') 
plt.title('Agression level depending on IQ') 
plt.xlabel('Agression') 
plt.ylabel('IQ') 
plt.show 
 
pearson_1 = data_1['Agression'].corr(data_1['IQ'], method = "pearson") 
spearman_1 = data_1['Agression'].corr(data_1['IQ'], method = "spearman") 
print(f"Pearson correlation: {round(pearson_1, 3)}") 
print(f"Pearson correlation: {round(spearman_1, 3)}")
```

```python
plt.plot(data_2['Solving speed'], data_2['IQ'], 'co') 
plt.title('Solving speed depending on IQ') 
plt.xlabel('Solving speed') 
plt.ylabel('IQ') 
plt.show 
 
pearson_2 = data_2['Solving speed'].corr(data_2['IQ'], method = "pearson") 
spearman_2 = data_2['Solving speed'].corr(data_2['IQ'], method = "spearman") 
print(f"Pearson correlation: {round(pearson_2, 3)}") 
print(f"Pearson correlation: {round(spearman_2, 3)}")
```
